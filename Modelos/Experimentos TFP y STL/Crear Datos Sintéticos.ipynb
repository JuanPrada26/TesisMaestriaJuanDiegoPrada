## AR1

```python
def AR1(phi,n,sigma,semilla):
    # phi: Coeficiente autoregresivo (autocorrelación)
    # n: Número de puntos en la serie
    # sigma: Desviación estándar del ruido

    # Inicialización de la serie
    np.random.seed(semilla) 
    X = np.zeros(n)    # Crear una serie de ceros
    epsilon = np.random.normal(0, sigma, size=n)  # Generar ruido blanco

    # Generar la serie AR(1)
    for t in range(1, n):
        X[t] = (phi * X[t-1]) + epsilon[t]
        
    return X
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Parámetros de la serie de tiempo
np.random.seed(0)
n_periods = 200  # Número de periodos
frecuencia = 24  # Frecuencia de la señal sinusoidal
amplitud = 10  # Amplitud de la onda
desplazamiento = 15  # Desplazamiento para mantener valores positivos
ruido = np.random.normal(0, 1, n_periods)  # Ruido blanco

# Generar datos sinusoidales
t = np.arange(n_periods)
data_sin = amplitud * np.sin(2 * np.pi * t / frecuencia) + ruido + desplazamiento

# Generar un componente estacional aleatorio
componente_estacional = np.random.normal(0, 2, n_periods)  # Componentes aleatorios
componente_estacional = np.sin(2 * np.pi * t / (frecuencia / 2)) * componente_estacional  # Frecuencia

# Sumar el componente estacional a la serie original
data_final = data_sin + componente_estacional

# Crear un DataFrame para la serie de tiempo
df = pd.DataFrame(data_final, columns=['valor'])

# Configurar y entrenar el modelo SARIMAX (opcional)
model = SARIMAX(df['valor'], order=(1, 0, 1), seasonal_order=(1, 1, 1, frecuencia))
sarimax_result = model.fit(disp=False)

# Graficar la serie generada
plt.figure(figsize=(10, 6))
plt.plot(df['valor'], label='Serie Original con Componente Estacional Aleatorio', color='blue')
plt.title('Serie de Tiempo Sintética con Comportamiento Sinusoidal y Componente Estacional Aleatorio')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Parámetros de la serie de tiempo
np.random.seed(0)
n_periods = 200  # Número de periodos
frecuencia = 24  # Frecuencia de la señal sinusoidal
amplitud = 10  # Amplitud de la onda
desplazamiento = 15  # Desplazamiento para mantener valores positivos
ruido = np.random.normal(0, 1, n_periods)  # Ruido blanco

# Generar datos sinusoidales
t = np.arange(n_periods)
data_sin = amplitud * np.sin(2 * np.pi * t / frecuencia) + ruido + desplazamiento

# Definir los períodos donde la serie se vuelve plana
plano_inicio = 50  # Inicio del periodo plano
plano_fin = 80     # Fin del periodo plano
valor_plano = np.mean(data_sin[plano_inicio:plano_fin])  # Valor constante durante el periodo

# Crear un nuevo array para la serie final
data_final = data_sin.copy()

# Aplicar el comportamiento plano
data_final[plano_inicio:plano_fin] = valor_plano

# Crear un DataFrame para la serie de tiempo
df = pd.DataFrame(data_final, columns=['valor'])

# Graficar la serie generada
plt.figure(figsize=(10, 6))
plt.plot(df['valor'], label='Serie Original con Comportamiento Sinusoidal y Período Plano', color='blue')
plt.axvspan(plano_inicio, plano_fin, color='red', alpha=0.3, label='Período Plano')  # Resaltar el período plano
plt.title('Serie de Tiempo Sintética con Comportamiento Sinusoidal y Período Plano')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Parámetros de la serie de tiempo
np.random.seed(0)
n_periods = 200
frecuencia = 24
amplitud = 10
desplazamiento = 15
ruido = np.random.normal(0, 1, n_periods)

# Generar datos cosenoidales
t = np.arange(n_periods)
data_cos = amplitud * np.cos(2 * np.pi * t / frecuencia) + ruido + desplazamiento

# Crear un nuevo array para la serie final
data_final = data_cos.copy()

# Definir el comportamiento plano y su duración
duracion_plana = 10
valor_plano = np.mean(data_cos)

# Repetir el comportamiento plano cada 24 periodos
for start in range(0, n_periods, 24):
    end = start + duracion_plana
    if end < n_periods:
        data_final[start:end] = valor_plano

# Crear un DataFrame para la serie de tiempo
df = pd.DataFrame(data_final, columns=['valor'])

# Ajustar un modelo SARIMA
order = (1, 0, 1)  # Parámetros ARIMA (p, d, q)
seasonal_order = (1, 0, 1, 24)  # Parámetros estacionales (P, D, Q, S)

model = sm.tsa.SARIMAX(df['valor'], order=order, seasonal_order=seasonal_order)
results = model.fit()

# Resumen del modelo ajustado
print(results.summary())

# Graficar los residuos
plt.figure(figsize=(10, 6))
plt.plot(results.resid)
plt.title('Residuos del Modelo SARIMA')
plt.axhline(0, color='red', linestyle='--')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Parámetros de la serie de tiempo
np.random.seed(0)
n_periods = 200
frecuencia = 24
amplitud = 10
desplazamiento = 15
ruido = np.random.normal(0, 1, n_periods)

# Generar datos cosenoidales
t = np.arange(n_periods)
data_cos = amplitud * np.cos(2 * np.pi * t / frecuencia) + ruido + desplazamiento

# Crear un nuevo array para la serie final
data_final = data_cos.copy()

# Definir el comportamiento plano y su duración
duracion_plana = 10
valor_plano = np.mean(data_cos)

# Repetir el comportamiento plano cada 24 periodos
for start in range(0, n_periods, 24):
    end = start + duracion_plana
    if end < n_periods:
        data_final[start:end] = valor_plano

# Crear un DataFrame para la serie de tiempo
df = pd.DataFrame(data_final, columns=['valor'])

# Ajustar un modelo SARIMA
order = (1, 0, 1)  # Parámetros ARIMA (p, d, q)
seasonal_order = (1, 0, 1, 24)  # Parámetros estacionales (P, D, Q, S)

model = sm.tsa.SARIMAX(df['valor'], order=order, seasonal_order=seasonal_order)
results = model.fit()

# Graficar la serie original y las predicciones
plt.figure(figsize=(12, 6))

# Graficar la serie original
plt.subplot(2, 1, 1)
plt.plot(df['valor'], label='Serie de Tiempo Original', color='blue')
plt.title('Serie de Tiempo Original')
plt.xlabel('Tiempo')
plt.ylabel('Valor')
plt.legend()

# Hacer predicciones
pred_start = len(df) - 10  # Comenzar las predicciones desde el penúltimo punto
pred = results.get_forecast(steps=10)
pred_index = np.arange(pred_start, pred_start + 10)
pred_mean = pred.predicted_mean
pred_ci = pred.conf_int()

# Graficar las predicciones
plt.subplot(2, 1, 2)
plt.plot(df['valor'], label='Serie de Tiempo Original', color='blue')
plt.plot(pred_index, pred_mean, label='Predicciones SARIMA', color='orange')
plt.fill_between(pred_index, 
                 pred_ci.iloc[:, 0], 
                 pred_ci.iloc[:, 1], 
                 color='pink', 
                 alpha=0.3, 
                 label='Intervalo de Confianza')
plt.title('Predicciones del Modelo SARIMA')
plt.xlabel('Tiempo')
plt.ylabel('Valor')
plt.legend()

plt.tight_layout()
plt.show()
```
